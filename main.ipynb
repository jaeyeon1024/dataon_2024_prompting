```python
import json
import pandas as pd
import openai
import open_key
import csv

# CSV 파일 경로 설정
csv_file_path = './docs/test.csv'

# 파일을 읽기 모드로 열기
with open(csv_file_path, mode='r', encoding='utf-8') as file:
    # CSV 파일을 읽을 준비
    reader = csv.reader(file)

    cnt = -1
    # 한 줄씩 읽어서 출력
    for row in reader:
        cnt += 1
        if cnt == 0:
            continue
        print(row)
        break


# OpenAI API 키를 설정합니다.


# 문단을 받아 질문과 답변 쌍을 생성하는 함수


data = pd.read_csv(csv_file_path)

data_for_llm = data[['ID', 'lang', 'title', 'notes']]

print(data_for_llm)
print(len(data_for_llm))


user_prompt = f"""

You will receive a CSV file where each row consists of ID, lang, title, and notes. Your task is to analyze the id, lang, title, and notes fields and determine whether they are related to automobiles or topics associated with automobiles.

The data will include entries from TEST_00 to TEST_39. For each TEST, you must output the result (either '1' for related or '0' for not related) on a new line.

1. Check if the number of lines in the response is 40.
2. If not, reconsider and rethink.

don't forget the TEST_39

\n {data_for_llm} """

system_prompt = f"""
You are an AI model that analyzes the given title and notices to determine if they are related to automobiles or topics related to automobiles.

The 'title' and 'notices' are independent columns, and the 'lang' column indicates the language of the text.

Process : 
0. Count the number of rows and remember it.
1. Think about topics related to automobiles, such as cars, traffic accidents, transportation, roads, etc.
2. Consider the key keywords and main topic of the notices.
3. Determine if the main topic, any keywords, title, or notices are associated with automobiles by considering topics such as cars, traffic accidents, transportation, roads, or other related subjects (e.g., car parts, traffic, driving, accidents, etc.).
4. Return '1' if related to automobiles, and '0' if not related.
5. Evaluate all rows and output '0' or '1' for each row based on the judgment.

Provide no additional information or text beyond '0' or '1' 
"""

#Provide no additional information or text beyond '0' or '1' 

question_response = openai.chat.completions.create(
    temperature=0.4,
    model="gpt-3.5-turbo-0125",
    
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]
)
question = question_response.choices[0].message.content.strip()


```
